```python
import datetime
import csv
import random

class FlightBookingSystem:
    def __init__(self):
        self.flights = []  # List to store flight details
        self.reservations = []  # List to store reservations
        self.passengers = []  # List to store passenger details

    def add_flight(self, flight_number, flight_name, airline, origin, destination, departure_time, seats_available, price, date):
        """Add a new flight to the system."""
        flight = {
            'flight_number': flight_number,
            'flight_name': flight_name,
            'airline': airline,
            'origin': origin,
            'destination': destination,
            'departure_time': departure_time,
            'seats_available': seats_available,
            'price': price,
            'date': date
        }
        self.flights.append(flight)
        print(f"Flight {flight_number} added successfully.")

    def find_flights(self, origin, destination):
        """Find available flights from origin to destination."""
        available_flights = [f for f in self.flights if f['origin'] == origin and f['destination'] == destination]
        if available_flights:
            print("Available flights:")
            for flight in available_flights:
                print(f"Flight {flight['flight_number']} ({flight['flight_name']}) by {flight['airline']} departs at {flight['departure_time']} on {flight['date']} with {flight['seats_available']} seats available for ${flight['price']}.")
        else:
            print("No available flights found.")
        return available_flights

    def reserve_seat(self, flight_number, passenger_name, passenger_email, passenger_phone):
        """Reserve a seat for a given flight number."""
        for flight in self.flights:
            if flight['flight_number'] == flight_number and flight['seats_available'] > 0:
                flight['seats_available'] -= 1
                reservation = {
                    'flight_number': flight_number,
                    'passenger_name': passenger_name,
                    'passenger_email': passenger_email,
                    'passenger_phone': passenger_phone,
                    'reservation_time': datetime.datetime.now()
                }
                self.reservations.append(reservation)
                self.passengers.append(reservation)
                print(f"Seat reserved for {passenger_name} on flight {flight_number}.")
                return reservation
        print("Reservation failed. Flight not found or no seats available.")
        return None

    def generate_passenger_report(self):
      """Generate a report of all passengers."""
      print("\nPassenger Report")
      print("=================")
      if not self.passengers:
          print("No passengers found.")
          return
      for passenger in self.passengers:
          print(f"Passenger: {passenger['passenger_name']}, Email: {passenger['passenger_email']}, Phone: {passenger['passenger_phone']}, Flight Number: {passenger['flight_number']}")
      print("=" * 40)

    def generate_reservation_report(self):
      """Generate a report of all reservations."""
      print("\nReservation Report")
      print("===================")
      if not self.reservations:
          print("No reservations made yet.")
          return
      for reservation in self.reservations:
          print(f"Passenger: {reservation['passenger_name']}, Flight: {reservation['flight_number']}, Reservation Time: {reservation['reservation_time']}")
      print("=" * 40)


    def save_flights_to_csv(self, filename="flights.csv"):
        """Save flight data to a CSV file."""
        with open(filename, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Flight Number", "Flight Name", "Airline", "Origin", "Destination", "Departure Time", "Seats Available", "Price", "Date"])
            for flight in self.flights:
                writer.writerow([flight['flight_number'], flight['flight_name'], flight['airline'], flight['origin'], flight['destination'], flight['departure_time'], flight['seats_available'], flight['price'], flight['date']])
        print(f"Flight data saved to {filename}.")

    def save_passengers_to_csv(self, filename="passengers.csv"):
        """Save passenger data to a CSV file."""
        with open(filename, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Passenger Name", "Email", "Phone", "Flight Number", "Reservation Time"])
            for passenger in self.passengers:
                writer.writerow([passenger['passenger_name'], passenger['passenger_email'], passenger['passenger_phone'], passenger['flight_number'], passenger['reservation_time']])
        print(f"Passenger data saved to {filename}.")

# Example usage
if __name__ == "__main__":
    system = FlightBookingSystem()
    airlines = ["American Airlines", "Delta Airlines", "United Airlines", "Southwest Airlines", "JetBlue"]
    cities = ["New York", "Los Angeles", "San Francisco", "Chicago", "Dallas", "Houston", "Miami", "Seattle", "Atlanta", "Denver"]

    for i in range(50):
        flight_number = f"AA{100 + i}"
        flight_name = f"Flight-{i + 1}"
        airline = random.choice(airlines)
        origin, destination = random.sample(cities, 2)
        departure_time = f"{random.randint(0, 23)}:{random.randint(0, 59):02d}"
        seats_available = random.randint(5, 50)
        price = random.randint(100, 1000)
        date = f"2025-02-{random.randint(10, 28)}"
        system.add_flight(flight_number, flight_name, airline, origin, destination, departure_time, seats_available, price, date)

    while True:
        print("\nFlight Booking System")
        print("1. Find Flights")
        print("2. Reserve a Seat")
        print("3. Generate Passenger Report")
        print("4. Generate Reservations Report")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            origin = input("Enter origin: ")
            destination = input("Enter destination: ")
            system.find_flights(origin, destination)
        elif choice == "2":
            flight_number = input("Enter flight number: ")
            passenger_name = input("Enter passenger name: ")
            passenger_email = input("Enter passenger email: ")
            passenger_phone = input("Enter passenger phone: ")
            system.reserve_seat(flight_number, passenger_name, passenger_email, passenger_phone)
        elif choice == "3":
            system.generate_passenger_report()
        elif choice == "4":
            system.generate_reservation_report()
        elif choice == "5":
            system.save_flights_to_csv()
            system.save_passengers_to_csv()
            print("Exiting system. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")
```

```python
import datetime
import csv
import random
import re

class FlightBookingSystem:
    def __init__(self):
        self.flights = []
        self.reservations = []
        self.passengers = []  # To store all passengers
        self.frequent_flyers = {}  # To track frequent flyer status

    def add_flight(self, flight_number, flight_name, airline, origin, destination, departure_time, seats_available, price, date):
        """Add a new flight to the system."""
        flight = {
            'flight_number': flight_number,
            'flight_name': flight_name,
            'airline': airline,
            'origin': origin,
            'destination': destination,
            'departure_time': departure_time,
            'seats_available': seats_available,
            'price': price,
            'date': date
        }
        self.flights.append(flight)

    def find_flights(self, origin, destination):
        """Find available flights from origin to destination."""
        origin = origin.strip().lower()
        destination = destination.strip().lower()

        print("\n🔍 Searching for flights...")

        available_flights = [
            f for f in self.flights
            if f['origin'].strip().lower() == origin and f['destination'].strip().lower() == destination
        ]

        if available_flights:
            print("\n✅ Available Flights:")
            for flight in available_flights:
                print(f"Flight {flight['flight_number']} ({flight['flight_name']}) by {flight['airline']}, "
                      f"departs at {flight['departure_time']} on {flight['date']} "
                      f"with {flight['seats_available']} seats available for ${flight['price']}.")
        else:
            print("\n❌ No available flights found. Please check the city names and try again.")

    def check_frequent_flyer_status(self, passenger_name):
        """Check the frequent flyer status based on bookings."""
        bookings = self.frequent_flyers.get(passenger_name, 0)
        if bookings >= 10:
            return "Gold"  # 10 or more bookings
        elif bookings >= 5:
            return "Silver"  # 5 or more bookings
        else:
            return "No Status"  # Less than 5 bookings

    def apply_discount(self, passenger_name, price):
        """Apply a discount based on the frequent flyer status."""
        status = self.check_frequent_flyer_status(passenger_name)
        discount = 0

        if status == "Gold":
            discount = 0.20  # 20% discount for Gold members
        elif status == "Silver":
            discount = 0.10  # 10% discount for Silver members

        if discount:
            print(f"\n🎉 {passenger_name} has a {int(discount*100)}% discount applied as a {status} member!")
            price -= price * discount
        return price

    def reserve_seat(self, flight_number, passenger_name, passenger_email, passenger_phone):
        """Reserve a seat for a given flight number and apply discounts for frequent flyers."""
        for flight in self.flights:
            if flight['flight_number'].lower() == flight_number.lower() and flight['seats_available'] > 0:
                flight['seats_available'] -= 1
                reservation = {
                    'flight_number': flight_number,
                    'passenger_name': passenger_name,
                    'passenger_email': passenger_email,
                    'passenger_phone': passenger_phone,
                    'reservation_time': datetime.datetime.now()
                }
                self.reservations.append(reservation)
                self.passengers.append(reservation)

                # Increment frequent flyer status for the passenger
                if passenger_name in self.frequent_flyers:
                    self.frequent_flyers[passenger_name] += 1
                else:
                    self.frequent_flyers[passenger_name] = 1

                # Apply discount based on frequent flyer status
                price_after_discount = self.apply_discount(passenger_name, flight['price'])
                print(f"\n✅ Seat reserved for {passenger_name} on flight {flight_number} at ${price_after_discount}.")
                return
        print("\n❌ Reservation failed. Flight not found or no seats available.")

    def generate_report(self, filename, report_type):
        """Generate a report from the CSV file."""
        print(f"\n📄 {report_type} Report (From CSV)")
        print("=" * 40)

        try:
            with open(filename, mode='r', newline='') as file:
                reader = csv.DictReader(file)
                data_found = False
                for row in reader:
                    data_found = True
                    print(", ".join(f"{key}: {value}" for key, value in row.items()))

                if not data_found:
                    print(f"No {report_type.lower()} found in the CSV file.")

        except FileNotFoundError:
            print(f"\n❌ The file {filename} was not found.")
        except Exception as e:
            print(f"\n⚠️ Error reading {filename}: {str(e)}")

        print("=" * 40)

    def get_valid_input(self, prompt, valid_options=None):
        """Get valid user input, ensuring it is not empty and within valid options."""
        while True:
            user_input = input(prompt).strip()
            if not user_input:
                print("❌ Input cannot be empty. Please try again.")
                continue
            if valid_options and user_input not in valid_options:
                print(f"❌ Invalid choice. Please enter one of {valid_options}.")
                continue
            return user_input

    def get_valid_email(self, prompt):
        """Validate email format."""
        while True:
            email = input(prompt).strip()
            if re.match(r"[^@]+@[^@]+\.[^@]+", email):
                return email
            print("❌ Invalid email format. Please enter a valid email.")

    def get_valid_phone(self, prompt):
        """Validate phone number to contain only digits."""
        while True:
            phone = input(prompt).strip()
            if phone.isdigit():
                return phone
            print("❌ Invalid phone number. Please enter digits only.")

    def load_passengers_from_csv(self, filename="passengers.csv"):
        """Load passenger reservations from a CSV file."""
        try:
            with open(filename, mode='r', newline='') as file:
                reader = csv.DictReader(file)

                # Ensure correct columns exist
                required_columns = {"Passenger Name", "Email", "Phone", "Flight Number", "Reservation Time"}
                if not required_columns.issubset(reader.fieldnames):
                    print("\n❌ CSV file is missing required columns.")
                    return

                self.passengers = []  # Clear existing passengers
                for row in reader:
                    self.passengers.append({
                        'passenger_name': row['Passenger Name'].strip().lower(),
                        'passenger_email': row['Email'],
                        'passenger_phone': row['Phone'],
                        'flight_number': row['Flight Number'],
                        'reservation_time': row['Reservation Time']
                    })

                print("\n✅ Passenger data loaded successfully from CSV.")
        except FileNotFoundError:
            print("\n❌ Error: Passenger CSV file not found.")
        except Exception as e:
            print(f"\n❌ An error occurred while reading passengers CSV: {e}")

    def find_passenger_details(self, passenger_name):
        """Find and display passenger reservation details from CSV-loaded data."""
        if not self.passengers:
            print("\n❌ No passenger data loaded. Please load passengers from CSV first.")
            return

        passenger_name = passenger_name.strip().lower()

        # Find the passenger (case-insensitive match)
        passenger = next((p for p in self.passengers if p['passenger_name'] == passenger_name), None)

        if passenger:
            print("\n📌 Passenger Reservation Details:")
            print("=" * 50)
            print(f"🔹 Name         : {passenger['passenger_name'].title()}")
            print(f"🔹 Email        : {passenger['passenger_email']}")
            print(f"🔹 Phone        : {passenger['passenger_phone']}")
            print(f"🔹 Flight No.   : {passenger['flight_number']}")
            print(f"🔹 Reservation  : {passenger['reservation_time']}")
            print("=" * 50)
        else:
            print("\n❌ No passenger found with this name.")

# Main program
if __name__ == "__main__":
    system = FlightBookingSystem()

    airlines = ["American Airlines", "Delta Airlines", "United Airlines", "Southwest Airlines", "JetBlue"]
    cities = ["New York", "Los Angeles", "San Francisco", "Chicago", "Dallas", "Houston", "Miami", "Seattle", "Atlanta", "Denver"]

    # Adding random flights
    for i in range(20):
        flight_number = f"AA{100 + i}"
        flight_name = f"Flight-{i + 1}"
        airline = random.choice(airlines)
        origin, destination = random.sample(cities, 2)
        departure_time = f"{random.randint(0, 23)}:{random.randint(0, 59):02d}"
        seats_available = random.randint(5, 50)
        price = random.randint(100, 1000)
        date = f"2025-02-{random.randint(10, 28)}"
        system.add_flight(flight_number, flight_name, airline, origin, destination, departure_time, seats_available, price, date)

    print("\nWelcome to the Flight Booking System! ✈️")

    # Ask if the user is an Admin or a Passenger
    user_type = system.get_valid_input("Are you an (1) Admin or (2) Passenger? Enter 1 or 2: ", valid_options=["1", "2"])

    if user_type == "2":  # Passenger Options
        while True:
            print("\nPassenger Menu:")
            print("1. Find Flights")
            print("2. Reserve a Seat")
            print("3. Flight Details")
            print("4. Check Discount")
            print("5. Exit")
            choice = system.get_valid_input("Enter your choice: ", valid_options=["1", "2", "3", "4", "5"])

            if choice == "1":
                origin = system.get_valid_input("Enter origin city: ")
                destination = system.get_valid_input("Enter destination city: ")
                system.find_flights(origin, destination)
            elif choice == "2":
                flight_number = system.get_valid_input("Enter flight number: ")
                passenger_name = system.get_valid_input("Enter passenger name: ")
                passenger_email = system.get_valid_email("Enter passenger email: ")
                passenger_phone = system.get_valid_phone("Enter passenger phone: ")
                system.reserve_seat(flight_number, passenger_name, passenger_email, passenger_phone)
            elif choice == "3":
                system.load_passengers_from_csv()
                passenger_name = system.get_valid_input("Enter passenger name: ")
                system.find_passenger_details(passenger_name)
            elif choice == "4":
                passenger_name = system.get_valid_input("Enter passenger name: ")
                system.check_frequent_flyer_status(passenger_name)
            elif choice == "5":
                print("\nThank you for using the system! Goodbye! 👋")
                break

    elif user_type == "1":  # Admin Options
        while True:
            print("\nAdmin Menu:")
            print("1. View Flights")
            print("2. Generate Reports")
            print("3. Exit")
            choice = system.get_valid_input("Enter your choice: ", valid_options=["1", "2", "3"])

            if choice == "1":
                for flight in system.flights:
                    print(f"Flight {flight['flight_number']} ({flight['flight_name']}) | {flight['airline']} | {flight['origin']} → {flight['destination']} | {flight['seats_available']} seats available | ${flight['price']} | {flight['date']}")

            elif choice == "2":
                print("\n1. Generate Passenger Report")
                print("2. Generate Reservation Report")
                report_choice = system.get_valid_input("Choose the report to generate: ", valid_options=["1", "2"])

                if report_choice == "1":
                    system.generate_report("passengers.csv", "Passenger")
                elif report_choice == "2":
                    system.generate_report("reservations.csv", "Reservation")

            elif choice == "3":
                print("\nExiting the system. Goodbye! 👋")
                break
```
